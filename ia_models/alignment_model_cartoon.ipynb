```python
from __future__ import print_function, division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# location to save figures
import os
cwd = os.getcwd()
figure_path = cwd + '/figures/'
```

```python
from matplotlib.patches import Circle, Ellipse, Arc
from matplotlib.collections import PatchCollection

fig, axes = plt.subplots(2,1, figsize=(3.3,7.0))
fig.subplots_adjust(left=0.00, right=1.0, bottom=0.1, top=0.9)

# left panel
ax = axes[0]
ax.set_title('with subhaloes')

halo = Ellipse((0.0,0.0),1.9, 1.5, color='grey', alpha=0.5)
ax.add_artist(halo)
halo_axis = Ellipse((0.0,0.0),1.9, 0.01, color='black', alpha=1)
ax.add_artist(halo_axis)

ax.text(-0.2, -0.7, 'host halo')
ax.text(-0.9, -0.1, 'halo axis')



subhalo = Ellipse((0.5,0.3),0.3, 0.5, color='grey', alpha=0.5, angle=-65)
ax.add_artist(subhalo)
subhalo_axis = Ellipse((0.5,0.3),0.01, 0.5, color='black', alpha=1, angle=-65)
ax.add_artist(subhalo_axis)

ax.text(0.1, 0.5, 'subhalo')

central = Ellipse((0.0,0.0),0.5, 0.25, color='red', alpha=0.5, angle=-30)
ax.add_artist(central)
central_axis = Ellipse((0.0,0.0),0.8, 0.01, color='black', alpha=1, angle=-30, linestyle='--')
ax.add_artist(central_axis)

ax.text(0.0, -0.4, 'central \n galaxy', color='red')

satellite = Ellipse((0.5,0.3),0.3, 0.15, color='red', alpha=0.5, angle=60)
ax.add_artist(satellite)
satellite_axis = Ellipse((0.5,0.3),0.5, 0.01, color='black', alpha=1, angle=60, linestyle='--')
ax.add_artist(satellite_axis)

ax.text(0.6, 0.1, 'satellite \n galaxy', color='red')


cen_host_angle = Arc((0.0,0.0), 0.6, 0.6, angle=0.0, theta1=150.0, theta2=180)
ax.add_artist(cen_host_angle)
ax.text(-0.5,0.1,r'$\theta_{\rm MA}$')

sat_sub_angle = Arc((0.5,0.3), 0.4, 0.4, angle=0.0, theta1=25, theta2=60)
ax.add_artist(sat_sub_angle)
ax.text(0.7,0.45,r'$\theta_{\rm MA}$')


ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.axis('off')

# right panel
ax = axes[1]
ax.set_title('without subhaloes')

halo = Ellipse((0.0,0.0),1.9, 1.5, color='grey', alpha=0.5)
ax.add_artist(halo)
halo_axis = Ellipse((0.0,0.0),1.9, 0.01, color='black', alpha=1)
ax.add_artist(halo_axis)

ax.text(-0.2, -0.7, 'host halo')
ax.text(-0.9, -0.1, 'halo axis')

central = Ellipse((0.0,0.0),0.5, 0.25, color='red', alpha=0.5, angle=-30)
ax.add_artist(central)
central_axis = Ellipse((0.0,0.0),0.8, 0.01, color='black', alpha=1, angle=-30, linestyle='--')
ax.add_artist(central_axis)

ax.text(0.0, -0.4, 'central \n galaxy', color='red')

satellite = Ellipse((0.5,0.3),0.3, 0.15, color='red', alpha=0.5, angle=70)
ax.add_artist(satellite)
satellite_axis = Ellipse((0.5,0.3),0.5, 0.01, color='black', alpha=1, angle=70, linestyle='--')
ax.add_artist(satellite_axis)

ax.text(0.6, 0.1, 'satellite \n galaxy', color='red')

ax.plot([0,0.5],[0,0.3],color='black')
ax.plot([0,0.5*1.5],[0,0.3*1.5],color='black')


cen_host_angle = Arc((0.0,0.0), 0.6, 0.6, angle=0.0, theta1=150.0, theta2=180)
ax.add_artist(cen_host_angle)
ax.text(-0.5,0.1,r'$\theta_{\rm MA}$')

sat_sub_angle = Arc((0.5,0.3), 0.4, 0.4, angle=0.0, theta1=30, theta2=70)
ax.add_artist(sat_sub_angle)
ax.text(0.65,0.5,r'$\theta_{\rm MA}$')

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.axis('off')

plt.show()

filename = 'model_cartoon.pdf'
fig.savefig(figure_path+filename, dpi=250)
```
