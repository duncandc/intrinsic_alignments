# Literature Models for IA

```python
from intrinsic_alignments.ia_models.literature_ia_model_components import Okumura09CentralAlignment
```

```python
ia_model_1 = Okumura09CentralAlignment()

theta_sample = np.linspace(-np.pi/2, np.pi/2.0, 1000)
pdf_1 = ia_model_1.misalignment_pdf(theta_sample)

ran_theta = ia_model_1.misalignment_rvs(1000000)

bins = np.linspace(0,1, 100)
counts = np.histogram(np.cos(ran_theta), bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure()
plt.step(bins[:-1], counts, where='post')
plt.show()


bins = np.linspace(-np.pi/2,np.pi/2,100)
counts = np.histogram(ran_theta, bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.plot(theta_sample, pdf_1)
plt.step(bins[:-1], counts, where='post')
plt.show()
```

```python
from intrinsic_alignments.ia_models.literature_ia_model_components import Bett12CentralAlignment
```

```python
ia_model_2 = Bett12CentralAlignment()
ia_model_2.param_dict['centrals_alingment_scatter'] = 0.8

costheta_sample = np.linspace(-1,1,1000)
pdf_2 = ia_model_2.misalignment_pdf(costheta_sample)

ran_theta = ia_model_2.misalignment_rvs(1000000)

bins = np.linspace(-1,1,100)
counts = np.histogram(np.cos(ran_theta), bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure()
plt.plot(costheta_sample, pdf_2)
plt.step(bins[:-1], counts, where='post')
plt.show()

bins = np.linspace(0,np.pi,100)
counts = np.histogram(ran_theta, bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure()
plt.step(bins[:-1], counts, where='post')
plt.plot(np.arccos(costheta_sample), pdf_2*np.sin(np.arccos(costheta_sample)))
plt.show()
```

```python
from intrinsic_alignments.ia_models.literature_ia_model_components import Knebe08SatelliteAlignment

ia_model_3 = Knebe08SatelliteAlignment()

costheta_sample = np.linspace(-1,1,1000)
pdf_3 = ia_model_3.misalignment_pdf(costheta_sample)

ran_theta = ia_model_3.misalignment_rvs(1000000)

bins = np.linspace(-1,1,100)
counts = np.histogram(np.cos(ran_theta), bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure()
plt.plot(costheta_sample, pdf_3*2)
plt.step(bins[:-1], counts, where='post')
plt.show()

bins = np.linspace(0,np.pi,100)
counts = np.histogram(ran_theta, bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure()
plt.plot(np.arccos(costheta_sample), pdf_3*np.sin(np.arccos(costheta_sample))*2)
plt.step(bins[:-1], counts, where='post')
plt.show()
```
