```python
from __future__ import print_function, division
import numpy as np
import matplotlib.pyplot as plt
from intrinsic_alignments.project_settings import PROJECT_DIRECTORY
%matplotlib inline
```

```python
from Illustris_Shapes.simulation_props import sim_prop_dict

simname = 'TNG300-1'
snapnum = 99
shape_type = 'iterative'
```

```python
from astropy.table import Table
fpath = fpath = PROJECT_DIRECTORY + 'modelling_illustris/data/'

sample_name = 'sample_1'
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_xi.dat'
result_1_xi = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_ed.dat'
result_1_ed = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_ee.dat'
result_1_ee = Table.read(fpath + fname, format='ascii')

sample_name = 'sample_2'
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_xi.dat'
result_2_xi = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_ed.dat'
result_2_ed = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_ee.dat'
result_2_ee = Table.read(fpath + fname, format='ascii')

sample_name = 'sample_3'
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_xi.dat'
result_3_xi = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_ed.dat'
result_3_ed = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_ee.dat'
result_3_ee = Table.read(fpath + fname, format='ascii')
```

```python
sample_name = 'sample_1'
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_xi.dat'
model_1_xi = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_ed.dat'
model_1_ed = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_ee.dat'
model_1_ee = Table.read(fpath + fname, format='ascii')

sample_name = 'sample_2'
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_xi.dat'
model_2_xi = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_ed.dat'
model_2_ed = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_ee.dat'
model_2_ee = Table.read(fpath + fname, format='ascii')

sample_name = 'sample_3'
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_xi.dat'
model_3_xi = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_ed.dat'
model_3_ed = Table.read(fpath + fname, format='ascii')
fname = simname + '_' + str(snapnum) + '-' + sample_name +'_model_ee.dat'
model_3_ee = Table.read(fpath + fname, format='ascii')
```

```python
fig, axes = plt.subplots(1,3, sharex=False, sharey=False, figsize=(8.5,3.3))
fig.subplots_adjust(left=0.1, right=0.95, bottom=0.2, top=0.9, wspace=0.35, hspace=0)

ax = axes[0]
p1 = ax.errorbar(result_1_xi['r'], result_1_xi['xi']*10, yerr=result_1_xi['err']*10,
            fmt= 'o', ms=4, color='blue')
l1 = ax.plot(model_1_xi['r'], model_1_xi['xi']*10,
             '-', color='blue')
p2 = ax.errorbar(result_2_xi['r'], result_2_xi['xi'], yerr=result_2_xi['err'],
            fmt= 'o', ms=4, color='orange')
l2 = ax.plot(model_2_xi['r'], model_2_xi['xi'],
             '-', color='orange')
p3 = ax.errorbar(result_3_xi['r'], result_3_xi['xi']/10, yerr=result_3_xi['err']/10,
            fmt= 'o', ms=4, color='green')
l3 = ax.plot(model_3_xi['r'], model_3_xi['xi']/10,
             '-', color='green')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim([0.01,100000])
ax.set_xlim([0.1,40])
ax.set_ylabel(r'$\xi(r)$')
ax.set_xlabel(r'$r ~[h^{-1}{\rm Mpc}]$')
ax.legend([p1,p2,p3],[r'$\log(M_*)\!\geq\!10.5$',r'$\log(M_*)\!\geq\!10.0$',r'$\log(M_*)\!\geq\!9.5$'],
           loc='upper right', frameon=False, fontsize=10, handletextpad=-0.5, title='TNG300-1')

ax = axes[1]
ax.errorbar(result_1_ed['r'], result_1_ed['ed'], yerr=result_1_ed['err'],
            fmt= 'o', ms=4, color='blue')
l1, = ax.plot(model_1_ed['r'], model_1_ed['ed'],
             '-', color='blue')
ax.fill_between(model_1_ed['r'], model_1_ed['ed']+model_1_ed['err'], model_1_ed['ed']-model_1_ed['err'],
                color='blue', alpha=0.25)
ax.errorbar(result_2_ed['r'], result_2_ed['ed'], yerr=result_2_ed['err'],
            fmt= 'o', ms=4, color='orange')
l2, = ax.plot(model_2_ed['r'], model_2_ed['ed'],
             '-', color='orange')
ax.fill_between(model_2_ed['r'], model_2_ed['ed']+model_2_ed['err'], model_2_ed['ed']-model_2_ed['err'],
                color='orange', alpha=0.25)
ax.errorbar(result_3_ed['r'], result_3_ed['ed'], yerr=result_3_ed['err'],
            fmt= 'o', ms=4, color='green')
l3, = ax.plot(model_3_ed['r'], model_3_ed['ed'],
             '-', color='green')
ax.fill_between(model_3_ed['r'], model_3_ed['ed']+model_3_ed['err'], model_3_ed['ed']-model_3_ed['err'],
                color='green', alpha=0.25)
ax.set_yscale('log')
ax.set_ylim([0.0002,0.8])
ax.set_xscale('log')
ax.set_xlim([0.1,40])
ax.set_ylabel(r'$\omega(r)$')
ax.set_xlabel(r'$r~[h^{-1}{\rm Mpc}]$')
ax.legend([l1,l2,l3],[r'$\log(M_*)\!\geq\!10.5$',r'$\log(M_*)\!\geq\!10.0$',r'$\log(M_*)\!\geq\!9.5$'],
           loc='upper right', frameon=False, fontsize=10, handletextpad=0.5, title='halo model')

ax = plt.subplot(333)
ax.plot([0.1,100],[0,0],'--', color='black')
l1 = ax.plot(model_1_ee['r'], model_1_ee['ee'],
             '-', color='blue')
ax.fill_between(model_1_ee['r'], model_1_ee['ee']+model_1_ee['err'], model_1_ee['ee']-model_1_ee['err'],
                color='blue', alpha=0.25)
ax.errorbar(result_1_ee['r'], result_1_ee['ee'], yerr=result_1_ee['err'],
            fmt= 'o', ms=4, color='blue')
ax.set_ylim([-0.05,0.05])
ax.set_yticks([-0.03,0.0,0.03])
ax.set_yticklabels([r'-3e-2','0','3e-2'])
ax.set_xscale('log')
ax.set_xlim([0.1,40])
ax.get_xaxis().set_visible(False)

ax = plt.subplot(336)
ax.plot([0.1,100],[0,0],'--', color='black')
ax.plot(model_2_ee['r'], model_2_ee['ee'],
        '-', color='orange')
ax.fill_between(model_2_ee['r'], model_2_ee['ee']+model_2_ee['err'], model_2_ee['ee']-model_2_ee['err'],
                color='orange', alpha=0.25)
ax.errorbar(result_2_ee['r'], result_2_ee['ee'], yerr=result_2_ee['err'],
            fmt= 'o', ms=4, color='orange')
ax.set_ylim([-0.005,0.005])
ax.set_yticks([-0.003,0.0,0.003])
ax.set_yticklabels(['-3e-3','0','3e-3'])
ax.set_xscale('log')
ax.set_xlim([0.1,40])
ax.set_ylabel(r'$\eta(r)$')
ax.get_xaxis().set_visible(False)

ax = plt.subplot(339)
ax.plot([0.1,100],[0,0],'--', color='black')
l1 = ax.plot(model_3_ee['r'], model_3_ee['ee'],
             '-', color='green')
ax.fill_between(model_3_ee['r'], model_3_ee['ee']+model_3_ee['err'], model_3_ee['ee']-model_3_ee['err'],
                color='green', alpha=0.25)
ax.errorbar(result_3_ee['r'], result_3_ee['ee'], yerr=result_3_ee['err'],
            fmt= 'o', ms=4, color='green')
ax.set_ylim([-0.005,0.005])
ax.set_yticks([-0.003,0.0,0.003])
ax.set_yticklabels(['-3e-3','0','3e-3'])
ax.set_xscale('log')
ax.set_xlim([0.1,40])
ax.set_xlabel(r'$r~[h^{-1}{\rm Mpc}]$')

plt.show()

fname = 'xi_edi_ee_plot.pdf'
fpath = PROJECT_DIRECTORY + 'modelling_illustris/figures/'
fig.savefig(fpath+fname, dpi=250)
```
