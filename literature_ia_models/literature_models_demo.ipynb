# Literature Models for IA

Here I test various models for galaxy-halo misalignments.  Diffferent models parameterize either the distribution of misalignment angle directly or the cosine of the misalignment angle.

```python
theta_sample = np.linspace(-np.pi/2.0, np.pi/2.0, 1000)
costheta_sample = np.linspace(-1,1,1000)
```

## Okumura (2009) Central Alignment Model

```python
from intrinsic_alignments.literature_ia_models.literature_ia_model_components import Okumura09CentralAlignment
```

```python
# sample from misalignment model

# load MA model
ia_model_1 = Okumura09CentralAlignment()

# analytical model
pdf_1 = ia_model_1.misalignment_pdf(theta_sample)

# rvs sample from model distribution
ran_theta = ia_model_1.misalignment_rvs(10000)

# binned in cos(theta)
bins = np.linspace(0,1,100)
counts = np.histogram(np.cos(ran_theta), bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure(figsize=(3.3,3.3))
l1, = plt.plot(np.cos(theta_sample), pdf_1/np.sin(theta_sample)*2.0, color='black')
l2, = plt.step(bins[:-1], counts, where='post', color='orange')
plt.xlabel(r'$\cos(\theta_{\rm MA})$')
plt.ylabel(r'$P(\cos(\theta_{\rm MA}))$')
plt.ylim([0,5])
plt.xlim([-0,1])
plt.legend([l1,l2],['model','random variates'], loc='upper left', frameon=False, fontsize=10)
plt.show()

# binned in theta
bins = np.linspace(-np.pi/2.0,np.pi/2.0,100)
counts = np.histogram(ran_theta, bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure(figsize=(3.3,3.3))
l1, = plt.plot(theta_sample, pdf_1, color='black')
l2, = plt.step(bins[:-1], counts, where='post', color='orange')
plt.xlabel(r'$\theta_{\rm MA}$')
plt.ylabel(r'$P(\theta_{\rm MA})$')
plt.ylim([0,1])
plt.xlim([-np.pi/2.0,np.pi/2.0])
plt.xticks([-np.pi/2.0,-np.pi/4.0,0,np.pi/4.0, np.pi/2.0],[r'$-\pi/2$',r'$-\pi/4$',r'$0$',r'$\pi/4$',r'$\pi/2$'])
plt.legend([l1,l2],['model','random variates'], loc='upper left', frameon=False, fontsize=10)
plt.show()
```

## Bett (2012) Central Alignment Model

```python
from intrinsic_alignments.literature_ia_models.literature_ia_model_components import Bett12CentralAlignment
```

```python
# load MA model
ia_model_2 = Bett12CentralAlignment()

# analytical model
pdf_2 = ia_model_2.misalignment_pdf(costheta_sample)

# rvs sample from model distribution
ran_theta = ia_model_2.misalignment_rvs(10000)

# binned in cos(theta)
bins = np.linspace(0,1,100)
counts = np.histogram(np.fabs(np.cos(ran_theta)), bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure(figsize=(3.3,3.3))
l1, = plt.plot(costheta_sample, pdf_2*2, color='black')
l2, = plt.step(bins[:-1], counts, where='post', color='orange')
plt.xlabel(r'$\cos(\theta_{\rm MA})$')
plt.ylabel(r'$P(\cos(\theta_{\rm MA}))$')
plt.ylim([0,5])
plt.xlim([0,1])
plt.legend([l1,l2],['model','random variates'], loc='upper left', frameon=False, fontsize=10)
plt.show()

# binned in theta
bins = np.linspace(-np.pi/2.0,np.pi/2.0,100)
counts = np.histogram(ran_theta, bins)[0]
counts = 1.0*counts/np.sum(counts)/np.diff(bins)

plt.figure(figsize=(3.3,3.3))
l1, = plt.step(bins[:-1], counts, where='post', color='orange')
l2, = plt.plot(np.arccos(costheta_sample), pdf_2*np.sin(np.arccos(costheta_sample)), color='black')
plt.plot(-1.0*np.arccos(costheta_sample), pdf_2*np.sin(np.arccos(costheta_sample)), color='black')
plt.xlabel(r'$\theta_{\rm MA}$')
plt.ylabel(r'$P(\theta_{\rm MA})$')
plt.ylim([0,1])
plt.xlim([-np.pi/2.0,np.pi/2.0])
plt.legend([l1,l2],['model','random variates'], loc='upper left', frameon=False, fontsize=10)
plt.show()
```
